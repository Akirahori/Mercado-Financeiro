**Teoria de Dow - Explicação Teórica**

Os conceitos da Teoria de Dow foi criada por Charles Dow por volta dos anos 1880. Ele publicava no *The Wall Street Journal* (jornal que ajudou a fundar) uma série de artigos com estudos do mercado. Charles Dow é considerado o pai da análise técnica e seus princípios vêm sendo usados há mais de 100 anos. Dow faleceu em 1902.

Após 20 anos de sua morte, ou seja, em 1922, William Hamilton reuniu todos esses artigos, categorizando e organizando-os, e publicou os princípios da Teoria. A partir disso, os princípios organizados da teoria de Dow começaram a se espalhar em Wall Street, e posteriormente, em todo o mundo.

O índice *Dow Jones Industrial Average* foi mais uma criação de Charles Dow. Ele é baseado na cotação de 30 das maiores e mais importantes empresas dos Estados Unidos e é o segundo índice para indicadores do mercado financeiro mais antigo dos Estados Unidos. 

**Os 6 princípios da Teoria de Dow**

**01) Os preços descontam tudo.**
  As ações, índices, commodities, etc, já possuem embutido em seus valores os eventos futuros que boa parte dos investidores desconhecem. As cotações já refletem o que o mercado acredita. Notícias já têm seu impacto incorporado aos valores dos ativos, exceto os chamados "atos de Deus".
  Segundo Dow, o operador deve realizar seus estudos e análises focando no movimento do próprio ativo, sem estudos paralelos. Tudo o que é importante já está descontado no preço.


**02) O Mercado tem três tendências.**

Charles Dow comparou a movimentação dos preços com os movimentos do mar. Ele percebeu a existência de três tendências ocorrendo ao mesmo tempo nos ativos: primária, secundária e terciária. 

**Tendência Primária:** principal tendência do mercado, movimentos duradouros, seja de alta ou baixa. Duração de meses até anos. 

**Tendência Secundária:** Movimentos de correção dentro da tendência primária. Elas são mais curtas que a tendência primária e ocorrem em sentido contrário a ela. Se a tendência primária é de alta, a secundária é de baixa, e vice-versa. Costumam corrigir de 1/3 a 2/3 da tendência primária precedente. Duração de 3 semanas até 3 meses. 

**Tendência Terciária:** Pequenas movimentações de preços que ocorrem dentro da tendência secundária. Oscilações diárias. Duração geralmente menor que 3 semanas. 

Muitas vezes não há uma distinção clara entre as tendências. Elas podem estar sobrepostas e cabe ao operador saber distinguí-las.

*Detalhe importante: na época de Charles Dow, era praticamente inexistente operações em Time Frames intradiários. As operações geralmente eram mais focadas ao longo prazo, de olho em fechamentos diários, semanais e mensais. Hoje, com o avanço da tecnologia, é possível operar time frames muito mais curtos e, com isso, esse conceito das três tendências pode ser adaptado a essa redução do tempo gráfico. Por exemplo, podemos considerar a tendência primária o gráfico de 60 minutos, a tendência secundária o gráfico de 5 minutos e a tendência terciária o grafico de 1 minuto. 
No entando, é importante salientar que essa adaptação não faz parte da teoria de Dow propriamente dita.*

*Essa adaptação (ou atualização) do princípio das três tendências do mercado gerou o que conhecemos como Fractal 5-3, onde toda a movimentação de preços é dividida em tendências menores, cada uma com 5 pernas a favor da tendência e 3 pernas de correção do movimento. Como se fosse uma peça encaixando na outra.*

**03) A tendências primárias possuem 3 fases.**

image.png


Quando a tendência primária é de **alta**, suas 3 fases são:

**Acumulação:** Acontece no início da tendência, pouco deslocamento do preço. É o momento antecipado ao resto do mercado. Difícil de prever porém o melhor momento para entrar em uma posição. Grupo de insiders e profissionais bem informados começam a comprar no momento em que o mercado está abaixo do valor justo. O público em geral ainda não se animou por completo.

**Subida sensível:** Momento onde a tendência já ficou clara para a maioria, portanto começam a adquirir os ativos. As informações se tornam publicas, a empresa começa a divulgar melhores resultados. Nessa fase, os preços tem uma movimentação significativa. Porém ainda nem todo o público está disposto a investir.

**Euforia:** Momento final e de maior amplitude de uma tendência. As notícias só falam dessa alta, sendo a hora ideal para profissionais sairem da posição. Justamente nesse momento, a maioria se convence que a tendência de alta é relevante e iniciam as compras. Mercado fica muito aquecido, fazem os preços subirem forte e rapidamente (movimento climático).

Quando a tendência primária é de **baixa**, suas 3 fases são:

**Distribuição:** Se inicia ao final da tendência de alta com os insiders e profissionais realizando lucro, ou seja, vendendo os ativos comprados no início de todo o movimento. Nessa fase, os preços geralmente ficam sem grandes movimentações, ou levemente em queda.

**Baixa Sensível:** Os compradores que entraram ao final do movimento começam a perder força e os vendedores prevalecem. Os investidores seguidores de tendência identificam a reversão e começam a vender, derrubando as cotações. Se inicia, ainda de forma pequena, os sentimentos e rumores negativos.

**Pânico:** Os sentimentos e rumores negativos ganham força, investidores querem se livrar do ativo com urgência, mídia atuando negativamente, o resto dos investidores começam a vender com o intuito de minimizar o prejuízo e encerrar suas posições. É um momento de desespero e queda acentuada.





**04) O volume deve confirmar a tendência.**

Conceito bem simples porém muito importante. Dow dizia que o volume deve confirmar a tendência vigente, seja de alta ou de baixa. Portanto, em uma tendência de alta, o volume deve aumentar nas pernas de alta e dimiuir durante as correções para baixo. Da mesma forma, em uma tendência de baixa, o volume deve aumentar durante as pernas de baixa e diminuir nas correções para cima. O volume é o que dá confiabilidade à tendência.

*Observe na imagem acima. O volume está na parte inferior do gráfico. As barras vermelhas são dos candles de baixa e as verdes dos candles de alta.  Temos uma tendência forte de baixa nesse gráfico.* 

*As maiores barras de volume ocorreram nos candles de baixa, confirmando a tendência. Já nas correções para cima o volume foi bem mais reduzido.*

*Perceba que não são em todas as barras de queda que o volume estava bem acima da média, porém é importante ter um bom senso e perceber que, durante todo esse movimento, as maiores barras de volume foram nos momentos de queda. Não tivemos nenhuma barra de volume verde tão grande quanto as maiores vermelhas.* 

**05) Princípio da confirmação**

No mercado acionário, há diversos índices que demonstram o desempenho de diferentes grupos de empresas. Para Dow, para que as tendências se confirmem, esses índices precisam se movimentar no mesmo sentido e intensidade. 

Por exemplo, se o S&P 500 rompe uma forte resistência e entra em forte tendência de alta, índice NASDAQ deve estar fazendo o mesmo movimento. Em uma eventual falha de um dos índices, o investidor deve ficar alerta a uma possível mudança na tendência principal.

Outro ponto importante é que na teoria de DOW  deve-se sempre usar o preço de fechamento nas análises. Os preços de abertura, máxima e mínima não são relevantes na teoria.
